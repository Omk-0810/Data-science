```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
x=np.array([2,13,19,32,38,45,58,65,71,88])
y=np.array([11,16,24,37,41,58,56,78,81,95])
z=np.array([81,80,79,63,54,32,29,26,18,14])
```

```python
plt.title('my first plot')
plt.xlabel('first value')
plt.ylabel('second value')
plt.grid()
plt.xlim(0,100)
plt.ylim(0,100)
plt.xticks(range(0,101,10))
plt.yticks(range(0,101,10))
plt.plot(x,z)
```

```python
plt.title('my first plot')
plt.xlabel('first value')
plt.ylabel('second value')
plt.grid()
plt.plot(x,y,color='red',marker='o',ls=':',mfc='green')
```

```python
plt.title('my first plot')
plt.xlabel('first value')
plt.ylabel('second value')
plt.grid()
plt.plot(x,y,color='red',label='2023')
plt.plot(x,z,color='green',label='2022')
plt.legend()
```

```python
plt.subplot(2,1,1)
plt.plot(x,y)
plt.subplot(2,1,2)
plt.plot(x,z)
```

```python
plt.subplot(121)
plt.plot(x,y)
plt.subplot(122)
plt.plot(x,z)
```

```python
plt.figure(figsize=(16,9))
plt.title('my first plot')
plt.xlabel('first value')
plt.ylabel('second value')
plt.grid()
plt.plot(x,z,lw=5,marker='o',ms=20,mfc='y')
plt.savefig('mypot.png')
```

#### Bar plot

```python
enrollment=[104,112,256,178,126,180]
courses=['c','Java','Python','R','PHP','Javascript']
```

```python
plt.xlabel('Courses')
plt.ylabel('Enrolled Students')
plt.grid()
plt.bar(courses,enrollment,color='brown')
```

```python
plt.xlabel('Courses')
plt.ylabel('Enrolled Students')
plt.grid()
plt.barh(courses,enrollment,color='brown')
```

### PIE CHART

```python
plt.figure(figsize=(10,10))
plt.pie(enrollment,labels=courses,autopct='%2.2f',shadow=True);
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('Salary_Data.csv')
```

```python
df.shape
```

```python
plt.scatter(df['YearsExperience'],df['Salary'])
```

```python
x=df['YearsExperience'].values
y=df['Salary']
```

```python
x
```

```python
y;
```

```python
x=x.reshape(30,1)
```

```python
x
```

#### import the linear regression class

```python
from sklearn.linear_model import LinearRegression
```

#### create the object of linear regression

```python
regressor= LinearRegression()
```

#### train the algorithm

```python
regressor.fit(x,y)
```

#### predict the unseen data

```python
regressor.predict([[15]])
```
